```python
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd

ticker_symbol = "TSLA"
tesla_data = yf.Ticker(ticker_symbol)
tesla_history = tesla_data.history(period="1d", start="2024-01-01", end="2024-05-01")
df_tesla_history=pd.DataFrame(tesla_history)


plt.figure(figsize=(10, 3))
plt.plot(tesla_history.index, tesla_history["Close"], label="Tesla (TSLA)", color='black')
plt.title("Tesla (TSLA) Stock Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.grid(True)
plt.legend()
plt.show()

```
